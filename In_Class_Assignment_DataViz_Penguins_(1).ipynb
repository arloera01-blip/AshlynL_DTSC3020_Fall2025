## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
# 1–3) Scatter plot with color by species
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette='Set2'
)
plt.title('Bill Length vs Bill Depth by Penguin Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

**Your interpretation (2–3 sentences):**

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
plt.figure(figsize=(8,5))
sns.histplot(data=penguins, x='bill_length_mm', bins=10, color='skyblue', kde=True)
plt.title('Bill Length Distribution (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(8,5))
sns.histplot(data=penguins, x='bill_length_mm', bins=30, color='lightcoral', kde=True)
plt.title('Bill Length Distribution (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._ In the chart with 30 bins, the count is higher while also having more variety between the bins. In the chart with 10 bins, more number are grouped togeather meaning, outliers are not as present as the one with more bins. However, the distrubution looks about the same as they both have a binomial shape.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
# Boxplot of body mass by species
plt.figure(figsize=(8,6))
sns.boxplot(data=penguins, x='species', y='body_mass_g', palette='pastel')
plt.title('Body Mass by Penguin Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()

```

**Your summary (1–2 sentences):**

_Write your observations here._ Typically the species Gentoo will be heavier compared to the other two species. The two species Adelie and Chinstrap will weigh about the same on average.
